```python
clientes, hoteis, avaliacoes, reservas, precos_disponibilidade, pagamentos, fidelidade = {}, {}, {}, {}, {}, {}, {}
id_cliente, id_hotel, id_reserva = 1, 1, 1

def selecionar_idioma():
    idiomas = {"1": "Português", "2": "Inglês", "3": "Espanhol"}
    print("\nEscolha o idioma:")
    for k, v in idiomas.items():
        print(f"{k} - {v}")
    opcao = input("Digite o número do idioma: ")
    return idiomas.get(opcao, "Português")

idioma_selecionado = selecionar_idioma()

def adicionar_cliente():
    global id_cliente
    nome = input("Nome: ")
    email = input("Email: ")
    telefone = input("Telefone: ")
    preferencias = input("Preferências de hospedagem: ")
    clientes[id_cliente] = {"Nome": nome, "Email": email, "Telefone": telefone, "Preferências": preferencias}
    fidelidade[id_cliente] = {"Pontos": 0, "Reservas": 0}
    print(f"Cliente {nome} cadastrado com ID {id_cliente}.\n")
    id_cliente += 1

def adicionar_hotel():
    global id_hotel
    nome = input("Nome do hotel: ")
    localizacao = input("Localização: ")
    quartos = int(input("Número de quartos disponíveis: "))
    preco = float(input("Preço por noite: "))
    hoteis[id_hotel] = {"Nome": nome, "Localização": localizacao, "Quartos": quartos}
    precos_disponibilidade[id_hotel] = {"Preço": preco, "Disponibilidade": quartos}
    print(f"Hotel {nome} cadastrado com ID {id_hotel}.\n")
    id_hotel += 1

def reservar_hotel():
    global id_reserva
    id_cliente = int(input("Seu ID de cliente: "))
    id_hotel = int(input("ID do hotel para reserva: "))
    if id_cliente not in clientes or id_hotel not in hoteis:
        print("Erro: Cliente ou hotel não encontrado.\n")
        return
    if precos_disponibilidade[id_hotel]["Disponibilidade"] == 0:
        print("Erro: Não há quartos disponíveis neste hotel.\n")
        return
    noites = int(input("Quantas noites deseja reservar? "))
    total = noites * precos_disponibilidade[id_hotel]["Preço"]
    print(f"Total a pagar: R${total:.2f}")
    pagamento(total)
    reservas[id_reserva] = {"Cliente": id_cliente, "Hotel": id_hotel, "Noites": noites, "Total": total}
    precos_disponibilidade[id_hotel]["Disponibilidade"] -= 1
    fidelidade[id_cliente]["Pontos"] += int(total / 10)
    fidelidade[id_cliente]["Reservas"] += 1
    print("Reserva realizada com sucesso!")
    id_reserva += 1

def modificar_reserva():
    id_reserva_mod = int(input("ID da reserva a modificar: "))
    if id_reserva_mod not in reservas:
        print("Erro: Reserva não encontrada.\n")
        return
    novas_noites = int(input("Novo número de noites: "))
    id_hotel = reservas[id_reserva_mod]["Hotel"]
    total_antigo = reservas[id_reserva_mod]["Total"]
    total_novo = novas_noites * precos_disponibilidade[id_hotel]["Preço"]
    print(f"Novo total: R${total_novo:.2f}")
    pagamento(total_novo - total_antigo)
    reservas[id_reserva_mod]["Noites"] = novas_noites
    reservas[id_reserva_mod]["Total"] = total_novo
    print("Reserva modificada com sucesso!")

def cancelar_reserva():
    id_reserva_del = int(input("ID da reserva a cancelar: "))
    if id_reserva_del in reservas:
        id_hotel = reservas[id_reserva_del]["Hotel"]
        precos_disponibilidade[id_hotel]["Disponibilidade"] += 1
        del reservas[id_reserva_del]
        print("Reserva cancelada com sucesso!")
    else:
        print("Erro: Reserva não encontrada.\n")

def pagamento(valor):
    print("Escolha a forma de pagamento:\n1 - Cartão de Crédito\n2 - Pix\n3 - Boleto")
    opcao = input("Digite a opção: ")
    print(f"Pagamento de R${valor:.2f} realizado com sucesso!")

def suporte_cliente():
    input("Digite sua dúvida sobre reservas: ")
    print("Nossa equipe entrará em contato em breve!\n")

def gerar_relatorios():
    print("\nRelatório de Reservas:")
    print(f"Total de reservas: {len(reservas)}")
    for id_hotel, info in hoteis.items():
        print(f"Hotel {info['Nome']} - Quartos disponíveis: {precos_disponibilidade[id_hotel]['Disponibilidade']}")

while True:
    print("\nMenu:")
    print("1 - Cadastrar Cliente")
    print("2 - Cadastrar Hotel")
    print("3 - Reservar Hotel")
    print("4 - Modificar Reserva")
    print("5 - Cancelar Reserva")
    print("6 - Consultar Fidelidade")
    print("7 - Suporte ao Cliente")
    print("8 - Gerar Relatórios")
    print("9 - Sair")
    opcao = input("Escolha uma opção: ")
    if opcao == "1":
        adicionar_cliente()
    elif opcao == "2":
        adicionar_hotel()
    elif opcao == "3":
        reservar_hotel()
    elif opcao == "4":
        modificar_reserva()
    elif opcao == "5":
        cancelar_reserva()
    elif opcao == "6":
        consultar_fidelidade()
    elif opcao == "7":
        suporte_cliente()
    elif opcao == "8":
        gerar_relatorios()
    elif opcao == "9":
        print("Saindo do sistema. Obrigado!")
        break
    else:
        print("Opção inválida. Tente novamente.\n")
```
